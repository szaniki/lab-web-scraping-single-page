# Standard scaling the features

```python
X = spotify_data[['acousticness', 'danceability', 'energy', 'instrumentalness', 'key','liveness','loudness', 'mode', 'speechiness',
           'tempo', 'valence', 'duration_ms','time_signature']]
```

```python
X.isna().sum() # no null values
```

```python
transformer = StandardScaler().fit(X)
X_prep = transformer.transform(X) 
pd.DataFrame(X_prep,columns=X.columns).head()


#import pickle
#pickle.dump(transformer, open('transformer.sav', 'wb'))
```

# Clustering

```python
%%time
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k') 
```

```python
%%time
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')

# the highest point shows me the opimal value for clusters aka 8
```

## The low elbow and high silhouette matching point suggests to work with 8 clusters

```python
kmeans = KMeans(n_clusters=8) #, random_state=1234) # defining the MODEL
kmeans.fit(X_prep)
```

```python
# MAGIC

#import pickle
#pickle.dump(kmeans, open('model.sav', 'wb'))
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```
